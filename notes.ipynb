SQL is a basic requiremnts for positions like Data Analyst, Data Engineer, Data Scienctist and Software Engineering/Development.

## Queries

### Select

The `SELECT` command is used to retrieve data from database


`SELECT col_name FROM table_name;`


we can use `*` in place of `col_name` to retrive all the columns

***NOTE:** SQL is NOT case-sensitive but, It is a convention to write the the sql keywords is Uppercase like SELECT, FROM, WHERE, etc* 

### LIMIT

Our above query will list down all the rows, if we want to list only 5 rows, we can add the limit clause at the end of our query like so.
`SELECT * FROM movies LIMIT 5;`

### OFFSET

It used to skip a specified number of rows from the beginning of a query's result

`SELECT * FROM movies OFFSET 5;`

### WHERE

We can even mention any conditions that we want using `WHERE` clause 

`SELECT * FROM movies WHERE <condition_1> AND actor_name="Robert";`

Also use  numeric conditions.

eg : `SELECT * FROM movies WHERE rating >= 9;`

we have following operators `>, >=, =, <, <=`

### AND, BETWEEN, OR, IS, IN, NULL 

These operators can be used with where clause

`SELECT * FROM movies WHERE rating BETWEEN 6 AND 9;`

`SELECT * FROM movies WHERE rating IS 6 OR 9 OR 7;`

`SELECT * FROM movies WHERE rating IS NULL;` 

replace the multiple OR operatores using a single IN operatore

`SELECT * FROM movies WHERE rating IN (6, 9, 7);`

### LIKE

The LIKE operator can be used within the WHERE clause to search for any specified patterns, 
eg: consider we have a table called 'movies' and a column called 'title' which contains the title of the movies and we want to search for all the movies containing the word "war",  we can do so using:

`SELECT * FROM movies WHERE title LIKE "%war%";`

We have specific characters for particular patters:
- % is a wild card character that matches 0 or more characters.

       eg: `a%`‑starts with a, `%a`‑ ends with a 

- _ is also wild card that matches only a single characters

       eg: `__s`‑ s is in second position

- [] matches single character within a specified range

       eg: `[a-z]`‑any char in range of a to z, `[0-9]`‑ any char from 0 to 9 


### Distinct

If we want to display only unique values we can use the distinct clause. Eg:

`SELECT DISTINCT <col_name> FROM <table_name>;`

### ORDER BY

we can use this clause to sort, we can use keywords `ASC`, `DESC` to order in ascending or descending order

`SELECT * FROM movies ORDER BY rating DESC`

### Functions

SQL functions are built-in operations or user-defined routines that perform calculations, manipulate data, and return results within a query, simplifying complex data handling tasks.

Types of Functions:
1. Built-In

    - Others

2. User Defined

### Group By

group rows that have the same values in specified columns into summary rows. It is almost always used with aggregate functions like COUNT(), SUM(), AVG(), MAX(), and MIN()

eg: 
    
    SELECT industry, AVG(rating) 

     FROM movies 

     GROUP BY industry;

### Alias

It is a temporary, alternative name given to a table or a column within a query to make it more readable and easier to write

`SELECT column_name AS alias_name FROM table_name;`

### Having

The `HAVING` clause is used to filter groups of records based on conditions involving aggregate functions (like COUNT(), SUM(), AVG(), etc.), after the data has been grouped by a GROUP BY clause. 




    SELECT release_year, COUNT(*) AS no_of_movies

    FROM movies
    
    GROUP BY release_year

    HAVING no_of_movies > 2

    ORDER BY no_of_movies DESC



***NOTE:** we cannot use the filter `no_of_movies > 2` within WHERE clause before GROUP BY, refer **Order of Execution.***

When using the HAVING clause we can only filter columns/aliases that we have defined within the SELECT clause, this is not the case with WHERE clause.

### Order of execution

`FROM and JOINS -> WHERE -> GROUP BY -> HAVING -> SELECT -> DISTINCT -> ORDER BY -> LIMIT and OFFSET`
