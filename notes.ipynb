# SQL

SQL is a language used to manage and query **Relational Databases**

A *relational database* is a database that organizes data into tables with rows and cols (like excel).

SQL is a basic requiremnts for positions like Data Analyst, Data Engineer, Data Scienctist and Software Engineering/Development.

## MySQL

MySQl is a RDBMS (relational database management system) software that uses SQL.

It is a open source software owned by Oracle.
